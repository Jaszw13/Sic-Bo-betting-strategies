````python
import pandas as pd
import random

# Assuming sic_bo_bets_df exists from previous step, if not, create a placeholder
if 'sic_bo_bets_df' not in locals() or sic_bo_bets_df.empty:
    sic_bo_bets_df = pd.DataFrame({
        'Bet Type': ['Small', 'Big'],
        'Description': ['Total of three dice is between 4 and 10 (inclusive), excluding triples.',
                        'Total of three dice is between 11 and 17 (inclusive), excluding triples.'],
        'Probability of Winning': [''] * 2,
        'Payout Ratio': [''] * 2,
        'House Edge (%)': [''] * 2,
        'Notes/Assumptions': [''] * 2
    })

# Populate 'Big' and 'Small' bet details based on standard Sic Bo rules
# There are 216 total outcomes (6^3)
# Winning outcomes for Small (4-10, excluding triples): 105 outcomes
# Winning outcomes for Big (11-17, excluding triples): 105 outcomes
# For simplicity, standard payout is 1:1 for Big/Small

small_prob = 105 / 216 # Probability of winning Small bet
big_prob = 105 / 216   # Probability of winning Big bet
payout_ratio_big_small = 1 # 1:1 payout

house_edge_small = (1 - (payout_ratio_big_small * small_prob)) * 100
house_edge_big = (1 - (payout_ratio_big_small * big_prob)) * 100

sic_bo_bets_df.loc[sic_bo_bets_df['Bet Type'] == 'Small', 'Probability of Winning'] = small_prob
sic_bo_bets_df.loc[sic_bo_bets_df['Bet Type'] == 'Small', 'Payout Ratio'] = payout_ratio_big_small
sic_bo_bets_df.loc[sic_bo_bets_df['Bet Type'] == 'Small', 'House Edge (%)'] = house_edge_small

sic_bo_bets_df.loc[sic_bo_bets_df['Bet Type'] == 'Big', 'Probability of Winning'] = big_prob
sic_bo_bets_df.loc[sic_bo_bets_df['Bet Type'] == 'Big', 'Payout Ratio'] = payout_ratio_big_small
sic_bo_bets_df.loc[sic_bo_bets_df['Bet Type'] == 'Big', 'House Edge (%)'] = house_edge_big

print("Updated sic_bo_bets_df with 'Big' and 'Small' bet details:")
print(sic_bo_bets_df[['Bet Type', 'Probability of Winning', 'Payout Ratio', 'House Edge (%)']])


def roll_dice():
    """Simulates rolling three standard six-sided dice."""
    return (random.randint(1, 6), random.randint(1, 6), random.randint(1, 6))

def check_win(bet_type, dice_results):
    """Checks if a given bet wins based on dice results, excluding triples for Big/Small."""
    d1, d2, d3 = dice_results
    total_sum = d1 + d2 + d3
    is_triple = (d1 == d2 == d3)

    if bet_type == 'Small':
        # Small bet wins if total is 4-10, and not a triple
        return (4 <= total_sum <= 10) and not is_triple
    elif bet_type == 'Big':
        # Big bet wins if total is 11-17, and not a triple
        return (11 <= total_sum <= 17) and not is_triple
    else:
        # Placeholder for other bet types if needed later
        raise ValueError(f"Unsupported bet type: {bet_type}")

def fixed_bet_strategy(initial_capital, target_capital, min_bet, max_bet, max_rounds, bet_type, sic_bo_bets_df_local):
    """Simulates a single run of the fixed-bet strategy.

    Args:
        initial_capital (int): Starting capital.
        target_capital (int): Capital to reach to win.
        min_bet (int): Minimum bet amount (fixed bet).
        max_bet (int): Maximum allowable bet (for clipping current_bet).
        max_rounds (int): Maximum number of rounds to play.
        bet_type (str): Type of bet ('Big' or 'Small').
        sic_bo_bets_df_local (pd.DataFrame): DataFrame containing bet details.

    Returns:
        tuple: (final_capital, rounds_played, capital_history)
    """
    current_capital = initial_capital
    current_bet = min_bet
    rounds_played = 0
    capital_history = [initial_capital]

    payout_ratio = sic_bo_bets_df_local.loc[sic_bo_bets_df_local['Bet Type'] == bet_type, 'Payout Ratio'].iloc[0]

    while 0 < current_capital < target_capital and rounds_played < max_rounds:
        rounds_played += 1

        # Ensure bet does not exceed current capital
        actual_bet = min(current_bet, current_capital)
        if actual_bet == 0:
            break # Cannot place a bet, essentially bankrupt

        current_capital -= actual_bet
        dice_results = roll_dice()

        if check_win(bet_type, dice_results):
            current_capital += actual_bet * (1 + payout_ratio)

        capital_history.append(current_capital)

    return current_capital, rounds_played, capital_history

def martingale_strategy(initial_capital, target_capital, min_bet, max_bet, max_rounds, bet_type, sic_bo_bets_df_local):
    """Simulates a single run of the Martingale strategy.

    Args:
        initial_capital (int): Starting capital.
        target_capital (int): Capital to reach to win.
        min_bet (int): Minimum bet amount (starting bet).
        max_bet (int): Maximum allowable bet.
        max_rounds (int): Maximum number of rounds to play.
        bet_type (str): Type of bet ('Big' or 'Small').
        sic_bo_bets_df_local (pd.DataFrame): DataFrame containing bet details.

    Returns:
        tuple: (final_capital, rounds_played, capital_history)
    """
    current_capital = initial_capital
    current_bet = min_bet
    rounds_played = 0
    capital_history = [initial_capital]

    payout_ratio = sic_bo_bets_df_local.loc[sic_bo_bets_df_local['Bet Type'] == bet_type, 'Payout Ratio'].iloc[0]

    while 0 < current_capital < target_capital and rounds_played < max_rounds:
        rounds_played += 1

        # Adjust current_bet for max_bet and available capital
        bet_to_place = min(current_bet, max_bet, current_capital)

        if bet_to_place <= 0: # Cannot place a bet, essentially bankrupt
            break

        current_capital -= bet_to_place
        dice_results = roll_dice()

        if check_win(bet_type, dice_results):
            current_capital += bet_to_place * (1 + payout_ratio)
            current_bet = min_bet # Reset bet on win
        else:
            current_bet *= 2 # Double bet on loss

        capital_history.append(current_capital)

    return current_capital, rounds_played, capital_history

print("Sic Bo game mechanics and betting strategy functions are defined.")
````
